# Drawing shapes in a canvas

```python
import numpy as np
```

```python
blank_img = np.zeros(shape = (512,512,3), dtype = np.int)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(img=blank_img,pt1=(200,100),pt2=(500,300),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(img=blank_img,pt1=(200,100),pt2=(500,300),color=(255,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center = (200,200), radius = 100, color = (0,255.255), thickness = 10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img, center = (50,50), radius = 10, color = (0,240,0), thickness = -1)
```

```python
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1 = (0,0), pt2 = (400,400),color = (190,240,0), thickness = 10)
```

```python
plt.imshow(blank_img)
```

# Write text

```python
font = cv2.FONT_ITALIC
```

### org is the bottom left coordinate of the base of the imaginary rectangle

```python
cv2.putText(blank_img, 'Ola! Amigo', org = (100,400), fontFace = font, fontScale = 4, color = (255,255,255), lineType = cv2.LINE_AA)
plt.imshow(blank_img)
```

# Drawing polygons

```python
blank_canvas = np.zeros((500,500,3), np.int)
```

```python
plt.imshow(blank_canvas)
```

```python
coords = np.array([[(100,400)],[(200,500)], [(400,400)], [(200,300)]], np.int)
```

```python
coords
```

```python
coords.shape
```

```python
cv2.polylines(blank_canvas, [coords], color=(0,255,0), thickness = 3, isClosed =True)
plt.imshow(blank_canvas)
```

# Drawing image with mouse

```python
import cv2
import numpy as np
def draw_circle(event, x,y, flags, params):
    if cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(blank_canvas, (x,y), radius = 10, color=(0,255,0), thickness = -1)
cv2.namedWindow(winname = 'canvas')
cv2.setMouseCallback('canvas', draw_circle)
blank_canvas = np.zeros((1080, 720, 3), np.int8)
while True:
    cv2.imshow('canvas', blank_canvas)
    if cv2.waitKey(1000) & 0XFF == 27:
        break
cv2.destroyAllWindows()
```
