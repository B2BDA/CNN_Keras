# Drawing image with mouse

1. First we create a function that we want to be done on input device execution
2. Then we want to make a window where we want to be executed
3. We connect the window with the callback

```python
import cv2
import numpy as np
def draw_circle(event, x,y, flags, params):
    if cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(blank_canvas, (x,y), radius = 10, color=(0,255,0), thickness = -1)
#         print(flags)
    elif cv2.EVENT_RBUTTONDOWN:
        cv2.cirle(blank_canvas, (x,y), radius  = 40, color = (0,0,255), thickness = -1)
#         print(flags)
cv2.namedWindow(winname = 'canvas')
cv2.setMouseCallback('canvas', draw_circle)
blank_canvas = np.zeros((1080, 720, 3), np.int8)
while True:
    cv2.imshow('canvas', blank_canvas)
    if cv2.waitKey(1000) & 0XFF == 27:
        break
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
draw = True
x1, y1 = (-1,-1)
def draw_rect(event, x,y, flags, params):
    global draw, x1, y1
    if event == cv2.EVENT_LBUTTONDOWN:
        x1,y1 = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        cv2.rectangle(blank_canvas, (x1,y1), (x,y), color = (0,0,255), thickness = -1)
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.circle(blank_canvas, (x1,y1), (x,y), color = (0,0,255), thickness = -1)

cv2.namedWindow(winname = 'boom.exe')
cv2.setMouseCallback('boom.exe', draw_rect)
blank_canvas = np.zeros((1080,720,3), np.int8)
while True:
    cv2.imshow('boom.exe', blank_canvas)
    if cv2.waitKey(10) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img_cv = cv2.imread(r"E:\Course\Python for Computer Vision with OpenCV and Deep Learning\1. Course Overview and Introduction\Computer-Vision-with-Python\DATA\dog_backpack.jpg",cv2.COLOR_BGR2RGB)

draw = False
x1,y1 = (-1,-1)
def circle(event, x,y, flags, params):
    global x1,y1
    if event == cv2.EVENT_LBUTTONDOWN:
        x1, y1 = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        cv2.circle(img_cv, center = (x1,y1), radius = 100, color = (0,0,0), thickness = 2)
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.circle(img_cv, center = (x1,y1), radius = 100, color = (0,0,0), thickness = 2)
cv2.namedWindow('doggo')
cv2.setMouseCallback('doggo', circle)

# image is the title of the window 
while True:
    cv2.imshow('doggo', img_cv)
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
        
        
```
